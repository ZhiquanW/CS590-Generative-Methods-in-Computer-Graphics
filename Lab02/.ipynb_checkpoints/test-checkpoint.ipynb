```python
# import packages
import numpy as np
import plotly.graph_objects as go
import scipy 
from bezier import *
# sample bezier
mcp_num = 12
sample_control_points = np.random.uniform(-10,10,size=(mcp_num,2))
sample_bezier = bezier(control_points=sample_control_points)
sample_bezier_points = sample_bezier.get_points(num=100)
# init bezier control points
cp_num = mcp_num
sp_num = 1000
control_points = np.random.uniform(-10,10,size=(cp_num,2))
control_points[0] = sample_bezier_points[0]
control_points[-1] = sample_bezier_points[-1]
bezier_curve = bezier(control_points=control_points)
bezier_samples = bezier_curve.get_points(num=sp_num)
print(cal_loss(sample_points=sample_bezier_points,line_points=bezier_samples))
# plot bezier curve & sample points
fig = go.Figure()
sample_bezier_points_scatter = go.Scatter(x=sample_bezier_points[:,0],y = sample_bezier_points[:,1],mode='lines+markers')
bezier_control_points_scatter = go.Scatter(x=control_points[:,0],y = control_points[:,1],mode='markers')
fig.add_traces([sample_bezier_points_scatter,bezier_control_points_scatter])
```

```python
# start optimization
from bezier import *
import matplotlib.pyplot as plt
max_iteration = 1000000
eps = 1
offset_range = np.array([-5,5])
offset_range_decay = 0.99
loss_recorder = []
total_loss = sys.float_info.max
for i in range(max_iteration):
    fig = go.Figure()
    sample_bezier_points_scatter = go.Scatter(x=sample_bezier_points[:,0],y = sample_bezier_points[:,1],mode='lines+markers')
    select_id = np.random.randint(1,cp_num-1)
    tmp_control_points = bezier_curve.get_control_points().copy()
    tmp_control_points[select_id] += np.random.uniform(*offset_range,size = 2)
    tmp_curve = bezier(control_points=tmp_control_points)
    tmp_loss = cal_loss(sample_points=sample_bezier_points,line_points=tmp_curve.get_points(num=sp_num))
    if tmp_loss < total_loss:
        offset_range *= offset_range_decay
        total_loss = tmp_loss
        loss_recorder.append(total_loss)
        print("iteration:",i," loss: " ,total_loss)
        bezier_curve.update_point(id = select_id,value=tmp_control_points[select_id])
        tmp_line_points = bezier_curve.get_points(num=sp_num)
        tmp_scatter = go.Scatter(x=tmp_line_points[:,0],y = tmp_line_points[:,1])
        bezier_control_points_scatter = go.Scatter(x=tmp_control_points[:,0],y = tmp_control_points[:,1],mode='markers')
        fig.add_traces([sample_bezier_points_scatter,tmp_scatter])
        fig.add_trace(bezier_control_points_scatter)
        fig.show()

print("Done")
```
